```python
from __future__ import annotations

from torch.utils.data import DataLoader

from model.config import BertConfig
from model.embedding import BertEmbeddings
from scripts.load_corpus import load_tokenized_corpus
from tokenizer.build_vocab import build_vocab
from training.dataset import MLMDataset


def main() -> None:
    vocab_size = 10_000
    max_seq_len = 128
    batch_size = 4

    tokenized_corpus = load_tokenized_corpus()
    vocab = build_vocab(
        tokenized_corpus=tokenized_corpus,
        vocab_size=vocab_size,
    )

    dataset = MLMDataset(
        tokenized_corpus=tokenized_corpus,
        vocab=vocab,
        max_seq_len=max_seq_len,
    )

    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=True,
    )

    batch = next(iter(dataloader))

    config = BertConfig(
        vocab_size=len(vocab),
        max_seq_len=max_seq_len,
        hidden_size=128,
        num_heads=4,
        num_layers=2,
        intermediate_size=512,
        dropout=0.1,
    )

    embeddings = BertEmbeddings(config)

    output = embeddings(batch["input_ids"])

    print("input_ids:", batch["input_ids"].shape)
    print("attention_mask:", batch["attention_mask"].shape)
    print("labels:", batch["labels"].shape)
    print("embedding output:", output.shape)


main()
```
